## practice

- 남자 승객의 결측치는 남자 승객의 평균
- 여자 승객의 결측치는 여자 승객의 평균
- 결측치를 모두 채우고 나서 모든 사람의 평균

```python
df_copy = df.copy()
```

```python
cond1 = df_copy['sex'] == 'male'
df_copy['age'].loc[cond1] = df_copy['age'].loc[cond1].fillna(df_copy.loc[cond1, 'age'].mean())
# 이렇게 하면 age의 시리즈에서 cond1을 찾는 것 loc를 먼저 해줘야함
```

```python
df_copy
```

```python
cond2 = df_copy['sex'] == 'female'
df_copy['age'].loc[cond2] = df_copy['age'].loc[cond2].fillna(df_copy.loc[cond2]['age'].mean())
```

```python
df_copy
```

```python
df_copy['age'].mean()
```

```python
df_copy = df.copy()
```

```python
cond1 = df_copy['sex'] == 'male'
```

```python
male_mean = df_copy.loc[cond1, 'age'].mean()
print(male_mean)
```

```python
cond2 = df_copy['sex'] == 'female'
```

```python
female_mean = df_copy.loc[cond2, 'age'].mean()
print(female_mean)
```

```python
df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_mean)
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(female_mean)
```

```python
df_copy['age'].mean()
```
