## practice

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.value_counts('day') # 데이터 형태 먼저 확인하기!
tips.value_counts('time')
tips.value_counts('size')
```

- 남성 흡연자 조건 필터링
- tip을 가장 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'

male_smoker = tips.loc[cond1 & cond2]
male_smoker.sort_values(['tip'], ascending=False).head(10)
```

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'
tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

- 식사 인원이 5명 이상인 저녁 테이블
- 흡연자가 없는 주말 테이블
- total_bill, tip 출력

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'
cond3 = tips['smoker'] == 'No'
cond4 = tips['day'] == 'Sun'
cond5 = tips['day'] == 'Sat'

my_tip = tips.loc[(cond1 & cond2 & cond3) & (cond4 | cond5)]
my_tip[['total_bill', 'tip']]
```

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(cond1 & cond2) & (cond3 & cond4)][['total_bill', 'tip']]
tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]
```
