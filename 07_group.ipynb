```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df_copy = df.copy()
```

## groupby()

```python
df_copy.groupby('sex').mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True) # 시리즈
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True) # [['survived']]: 예쁘게 보기
```

### reset_index()

```python
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived']].mean()
df_temp
```

```python
df_temp.reset_index()
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])
```

## pivot_table()

```python
df_copy.pivot_table(index='who', values='survived')
```

```python
df_copy.pivot_table(index=['who', 'pclass'], values='survived')
```
