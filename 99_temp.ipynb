```python
cond_a = dfc['species'] == 'Adelie'
cond_g = dfc['species'] == 'Gentoo'
cond_c = dfc['species'] == 'Chinstrap'

cond_m = dfc['sex'] == 'Male'
cond_f = dfc['sex'] == 'Female'

print(dfc.loc[(cond_a & cond_m), 'body_mass_g'].mean())
print(dfc.loc[(cond_a & cond_f), 'body_mass_g'].mean())
print(dfc.loc[(cond_g & cond_m), 'body_mass_g'].mean())
print(dfc.loc[(cond_g & cond_f), 'body_mass_g'].mean())
print(dfc.loc[(cond_c & cond_m), 'body_mass_g'].mean())
print(dfc.loc[(cond_c & cond_f), 'body_mass_g'].mean())
```

```python
dfc.sort_values('bill_depth_mm')
```

```python
cond_a = dfc['species'] == 'Adelie'
cond_g = dfc['species'] == 'Gentoo'

cond_t = dfc['island'] == 'Torgersen'
cond_b = dfc['island'] == 'Biscoe'

print(dfc.loc[(cond_a & cond_t), 'bill_depth_mm'].mean())
print(dfc.loc[(cond_g & cond_b), 'bill_depth_mm'].mean())

dfc_at = dfc.loc[(cond_a & cond_t), 'bill_depth_mm']
dfc_gb = dfc.loc[(cond_g & cond_b), 'bill_depth_mm']

at_mean = dfc.loc[(cond_a & cond_t), 'bill_depth_mm'].mean()
gb_mean = dfc.loc[(cond_g & cond_b), 'bill_depth_mm'].mean()



# dfc_at = dfc_at.fillna(at_mean)
dfc.loc[(cond_g & cond_b), 'bill_depth_mm'] = dfc_gb.fillna(gb_mean)

# dfc.sort_values('bill_depth_mm')
dfc
```

```python
dfc.loc[(cond_a & cond_t), 'bill_depth_mm'] = dfc.loc[(cond_a & cond_t), 'bill_depth_mm'].fillna(at_mean)
dfc.sort_values('bill_depth_mm')
```
